# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt

sql_query = """
SELECT ConvertedComp
FROM master
"""

# Fetch the data from the database into a pandas DataFrame
df = pd.read_sql_query(sql_query, conn)

# Plot the histogram of ConvertedComp
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=30, edgecolor='black')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
sql_query = """
SELECT Age 
FROM master
"""

# Fetch the data from the database into a pandas DataFrame
Age = pd.read_sql_query(sql_query, conn)

# Create a box plot of the Age column
plt.figure(figsize=(10, 8))
boxplot = Age.boxplot(column=['Age'], showfliers=True)  # Show outliers

# Calculate Q1, Q2 (median), Q3, and IQR
Q1 = Age['Age'].quantile(0.25)
Q2 = Age['Age'].quantile(0.50)  # Median
Q3 = Age['Age'].quantile(0.75)
IQR = Q3 - Q1

# Highlight outliers with different colors
outliers = Age[(Age['Age'] < (Q1 - 1.5 * IQR)) | (Age['Age'] > (Q3 + 1.5 * IQR))]
outlier_indices = outliers.index.tolist()

for i in outlier_indices:
    plt.scatter(x=1, y=Age.loc[i], color='red', marker='o', s=100, label='Outlier')

# Enhance the box plot to show Q1, Q2 (median), Q3, and IQR in different colors
plt.axhline(Q1, color='r', linestyle='--', label='Q1')  # Red dashed line for Q1
plt.axhline(Q2, color='g', linestyle='-', label='Q2 (Median)')  # Green solid line for Q2 (median)
plt.axhline(Q3, color='b', linestyle='--', label='Q3')  # Blue dashed line for Q3
plt.axhspan(Q1, Q3, facecolor='yellow', alpha=0.2, label='IQR')  # Yellow shaded region for IQR

# Add legend and labels
plt.legend()
plt.xlabel('Age')
plt.ylabel('Values')
plt.title('Box Plot of Age with Q1, Q2, Q3, IQR, and Outliers Highlighted')

plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
Query8 = """
SELECT Avg(WorkWeekHrs), Avg(CodeRevHrs), Age 
FROM master 
where Age between 30 and 35
group by Age
"""

z = pd.read_sql_query(Query8, conn)
WorkWeekHrs= z['Avg(WorkWeekHrs)']
CodeRevHrs= z['Avg(CodeRevHrs)']
Age= z['Age']

fig, ax = plt.subplots()

ax.bar(Age, WorkWeekHrs, label='WorkWeekHrs')
ax.bar(Age, CodeRevHrs, bottom=WorkWeekHrs,
       label='CodeRevHrs')

plt.show()
```

```python
sql_query = '''
SELECT DatabaseDesireNextYear, COUNT(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
'''

# Fetch the data from the database into a pandas DataFrame
data = pd.read_sql_query(sql_query, conn)

# Calculate the percentages for each database
data['Percentage'] = (data['Count'] / data['Count'].sum()) * 100

# Display the resulting DataFrame
print("Top 5 Databases Respondents Wish to Learn Next Year:")
print(data)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(data['Percentage'], labels=data['DatabaseDesireNextYear'], autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

```python
# Define your SQL query to get the count of respondents interested in each language
sql_query = '''
SELECT LanguageDesireNextYear, COUNT(*) as Count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC
'''

# Fetch the data from the database into a pandas DataFrame
data = pd.read_sql_query(sql_query, conn)

# Calculate the rank of each language based on popularity
data['Rank'] = data['Count'].rank(ascending=False, method='min')

# Display the resulting DataFrame with rank
print("Rank of Languages Respondents Wish to Learn Next Year:")
print(data)

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.barh(data['LanguageDesireNextYear'], data['Count'])
plt.xlabel('Number of Respondents')
plt.ylabel('Languages')
plt.title('Popularity of Languages Respondents Wish to Learn Next Year')
plt.gca().invert_yaxis()  # Invert y-axis to display the most popular language at the top
plt.show()
```

```python
sql_query = '''
SELECT COUNT(*) as Count
FROM DatabaseWorkedWith
'''

# Fetch the data from the database into a pandas DataFrame
data = pd.read_sql_query(sql_query, conn)

# Display the count of respondents working with databases
print("Number of Respondents Currently Working with Databases:")
print(data['Count'][0])

# Plotting the graph
plt.figure(figsize=(6, 4))
plt.bar(['Respondents'], data['Count'])
plt.xlabel('Respondents')
plt.ylabel('Count')
plt.title('Respondents Currently Working with Databases')
plt.show()
```

```python


# Define SQL queries to get the count of respondents working with each database and specific databases
sql_query_all_databases = '''
SELECT DatabaseWorkedWith, COUNT(*) as Count
FROM DatabaseWorkedWith
GROUP BY DatabaseWorkedWith
ORDER BY Count DESC
'''

sql_query_sql = '''
SELECT COUNT(*) as Count
FROM DatabaseWorkedWith
WHERE DatabaseWorkedWith LIKE '%SQL%'
'''

sql_query_mysql_only = '''
SELECT COUNT(*) as Count
FROM DatabaseWorkedWith
WHERE DatabaseWorkedWith = 'MySQL'
'''

# Fetch the data from the database into pandas DataFrames
data_all_databases = pd.read_sql_query(sql_query_all_databases, conn)
data_sql = pd.read_sql_query(sql_query_sql, conn)
data_mysql_only = pd.read_sql_query(sql_query_mysql_only, conn)

# Display the count of respondents working with each database
print("Number of Respondents Working with Each Database:")
print(data_all_databases)

# Display the count of respondents working with 'SQL' and 'MySQL' only
print("\nNumber of Respondents Working with 'SQL':")
print(data_sql['Count'][0])

print("\nNumber of Respondents Working with 'MySQL' Only:")
print(data_mysql_only['Count'][0])

# Plotting the graph for all databases
plt.figure(figsize=(10, 6))
plt.bar(data_all_databases['DatabaseWorkedWith'], data_all_databases['Count'])
plt.xlabel('Database')
plt.ylabel('Count')
plt.title('Respondents Working with Each Database')
plt.xticks(rotation=45, ha='right')
plt.show()
```
